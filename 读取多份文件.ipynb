```python
#
'''
同一个文件夹下，有很多要处理的json文件，如果这些都是需要一起拿来用，那么下面的demo和思路可以帮助到你

主要是借助os.walk 和os.path.join的功能

os.walk传入原始路径，提供遍历每个文件的功能
os.path.join传入原始路径(parents,)和文件名(filename,)

'''

import os
fs_mention_entity = pd.DataFrame()  #为空，默认列名为要读的文件列名。
inputpath = '../data/fs_mention_pair/'

for parents, dirnames, filenames in os.walk(inputpath):
    for filename in filenames:
        df = pd.read_json(os.path.join(parents, filename),orient='records', lines = True)
        fs_mention_entity = fs_mention_entity.append(df, ignore_index = True)

```
